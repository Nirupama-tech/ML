```python
data=pd.read_csv("train.csv")

```

```python
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Attrition",y="Number of Promotions",data=data)
plt.title("Number of promotions for people who stayed vs left")
plt.xlabel("Attrition")
plt.ylabel("Number of Promotions")
plt.show()
stayed=data[data["Attrition"]=="Stayed"]["Number of Promotions"]
left=data[data["Attrition"]=="Left"]["Number of Promotions"]
t_stat, p_value=stats.ttest_ind(stayed, left, nan_policy='omit')
print(f"TStatistics:{t_stat:.2f}")
print(f"P-Value:{p_value:.2f}")
if(p_value<0.05):
  print("There is statistically significant relation between number of promotions and attrition status")
else:
  print("There is no significant relation between number of promotions and attrition status")
```
