```python
data=pd.read_csv("train.csv")
data.head()
```

```python
count_left=data[data['Attrition']=='Left'].groupby('Overtime').size()
total_count_overtime=data.groupby('Overtime').size()
percentage= (count_left/total_count_overtime)*100
def plot_pie_chart(data,title):
  if data.empty:
    print("No data to plot")
  else:
    total=data.sum()
    plt.figure(figsize=(8,6))
    plt.pie(
    data,
    labels=data.index,
    autopct=lambda pct: '{:.1f}%'.format(pct * total / 100),  # Format percentages
    colors=plt.cm.Paired(range(len(data))))
    plt.title(title)
    plt.show()
plot_pie_chart(percentage,'Percentages of Resignations by Overtime Category')

data[data['Attrition']=='Stayed']['Overtime']

```
